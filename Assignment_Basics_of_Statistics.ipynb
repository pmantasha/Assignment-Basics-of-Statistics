#Basics of Statistics:

#Theory:-

1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.

  -  Types of Data

    - Qualitative Data: Non-numeric, describes categories or characteristics.

        Example: Eye color (brown, blue), brands of cars.

    - Quantitative Data: Numeric, measures quantities.

        Example: Heights, ages, temperatures.

  -  Scales:

    - Nominal: Categories without order (e.g., gender, nationality).

    - Ordinal: Categories with a meaningful order (e.g., satisfaction ratings).

    - Interval: Numeric scales with equal intervals but no true zero (e.g., temperature in Celsius).

    - Ratio: Numeric scales with a true zero (e.g., weight, height).


2. What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.

   - Measures of Central Tendency
     - Mean: Average of all data points. Best for symmetric distributions without outliers.

     - Median: Middle value when data is ordered. Best for skewed distributions or data with outliers.

     - Mode: Most frequently occurring value. Useful for categorical data.

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

    - Dispersion
      - Describes how spread out the data is.

      - Variance: Average of squared deviations from the mean.

      - Standard Deviation: Square root of variance; easier to interpret as it's in the same unit as the data.

4. What is a box plot, and what can it tell you about the distribution of data?

    - Box Plot

      - Visual summary showing median, quartiles, minimum, maximum, and outliers.

      - Helps detect skewness and variability.

5. Discuss the role of random sampling in making inferences about populations.

    - Random Sampling

     - Each member of the population has an equal chance of being selected.

     - Ensures unbiased, representative samples for valid inferences about populations.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

   - Skewness:
     Measures asymmetry in data distribution.

     Types:

     - Positive Skew: Tail extends to the right.  

     - Negative Skew: Tail extends to the left.

     Affects interpretation: the mean is pulled toward the skew.

7. What is the interquartile range (IQR), and how is it used to detect outliers?

    - Interquartile Range (IQR)

     - Difference between the 75th percentile (Q3) and 25th percentile (Q1).

     - Used to detect outliers: data points 1.5 × IQR above Q3 or below Q1 are typically considered outliers.

8. Discuss the conditions under which the binomial distribution is used.

   - Binomial Distribution Conditions

    - Fixed number of trials.

    - Only two possible outcomes (success or failure).

    - Constant probability of success.

    - Independent trials.

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

    - Normal Distribution & Empirical Rule

     Symmetrical, bell-shaped curve.

     Empirical Rule:

     - 68% within ±1 standard deviation.

     - 95% within ±2 standard deviations.

     - 99.7% within ±3 standard deviations.

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

    - Poisson Process Example

     - Example: Number of customer arrivals at a store per hour.

     - Poisson distribution models the probability of a given number of events happening in a fixed interval when events occur independently and at a constant average rate.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

    - Random Variables

     - Random Variable: Numerical outcome of a random process.

     - Discrete: Countable outcomes (e.g., number of heads in coin flips).

     - Continuous: Infinite possible values within a range (e.g., height, weight).

12. Provide an example dataset, calculate both covariance and correlation, and interpret the results.

  - Covariance and Correlation Example

    - Dataset Example:
    X=[1,2,3,4,5], Y=[2,4,6,8,10]

    - Covariance measures the direction of the relationship.

    - Correlation measures both strength and direction, ranging from -1 to 1.

    - Positive correlation means as one variable increases, the other tends to increase.
